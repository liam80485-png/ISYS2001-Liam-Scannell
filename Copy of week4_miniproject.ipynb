## Part 1: Enhanced Expense Entry (15 minutes)

### 🔄 From One to Many

Transform your Week 2 single-expense tracker to handle multiple expenses with proper validation:

```python
# 1. Start the program
print("Welcome to the Finance Tracker App!")

# Initialize lists and variables for expenses
expenses = []
total_spent = 0.0

# 2. Prompt for user's name
name = input("Enter your name: ")

# 3. Prompt for preferred theme color
theme_color = input("Enter your preferred theme color: ")

# 4. Prompt for primary currency
currency = input("Enter your primary currency (e.g., USD, AUD, EUR): ")

# 5. Prompt for main savings goal
savings_goal = input("Enter your main savings goal: ")

# 6. Prompt for notification preference
notification_preference = input("Do you want to receive daily notifications? (yes/no): ")

# 7. Prompt for daily budget with validation
daily_budget = 0.0
while True:
    try:
        daily_budget_str = input("What is your preferred daily budget?: ")
        daily_budget = float(daily_budget_str)
        if daily_budget <= 0:
            print("Daily budget must be a positive number. Please try again.")
        else:
            break
    except ValueError:
        print("Invalid input. Please enter a numerical value for your daily budget.")

# Assume a monthly budget for analysis in Part 2
# (adjust as needed based on project requirements)
monthly_budget = daily_budget * 30 # Simple approximation for a month

# 8. Construct welcome and summary message
summary_message = (
    f"\nHello {name}!\n"
    f"Your app theme color is set to {theme_color}.\n"
    f"Your primary currency is {currency}.\n"
    f"Your main savings goal is: {savings_goal}.\n"
    f"Daily notifications are turned {notification_preference}.\n"
    f"Your preferred daily budget is: {daily_budget:.2f} {currency} (Monthly: {monthly_budget:.2f} {currency}).\n"
    f"You're all set to start tracking your finances!"
)

# 9. Print the summary message
print(summary_message)


# Part 1: Enhanced Expense Entry
print("\n=== Expense Entry ===")
print("Enter expenses one by one. Type 'done' for amount when finished.")

while True:
    expense_amount_str = input("Enter expense amount (or 'done' to finish): ").lower()
    if expense_amount_str == 'done':
        break

    try:
        amount = float(expense_amount_str)
        if amount <= 0:
            print("Expense amount must be a positive number. Please try again.")
            continue
    except ValueError:
        print("Invalid input. Please enter a numerical value for the amount.")
        continue

    description = input("Enter expense description: ")
    category = input("Enter expense category (e.g., Food, Transport, Utilities): ") # Optional: can add more validation here

    expenses.append({
        'amount': amount,
        'description': description,
        'category': category
    })
    total_spent += amount
    print(f"Added: {description} - {currency} {amount:.2f}")

print(f"\nTotal expenses entered: {len(expenses)}")
print(f"Total spent so far: {currency} {total_spent:.2f}")

# 10. End the program (initial setup part)
print("\nSetup complete and expenses entered. Thank you for using the Finance Tracker App!")
```

## Part 2: Basic Expense Analysis (15 minutes)

### 📈 Processing Your Transaction Data

Use loops to analyze the expenses you collected:

```python
# Expense Analysis Section
print("\n=== Expense Analysis ===")

if len(expenses) > 0:
    # Basic statistics
    print(f"Total expenses recorded: {len(expenses)}")
    print(f"Total amount spent: {currency} {total_spent:.2f}")

    # Calculate average expense
    average_expense = total_spent / len(expenses)
    print(f"Average expense: {currency} {average_expense:.2f}")

    # Budget usage
    budget_used_percentage = (total_spent / monthly_budget) * 100 if monthly_budget > 0 else 0
    remaining_budget = monthly_budget - total_spent
    print(f"Monthly Budget Used: {budget_used_percentage:.1f}%")
    print(f"Remaining Monthly Budget: {currency} {remaining_budget:.2f}")

    # Find highest and lowest expenses
    highest_expense = expenses[0]['amount']
    lowest_expense = expenses[0]['amount']
    highest_description = expenses[0]['description']
    lowest_description = expenses[0]['description']

    for expense in expenses:
        if expense['amount'] > highest_expense:
            highest_expense = expense['amount']
            highest_description = expense['description']
        if expense['amount'] < lowest_expense:
            lowest_expense = expense['amount']
            lowest_description = expense['description']

    print(f"\n💰 Expense Range:")
    print(f"Highest expense: {currency} {highest_expense:.2f} (Description: {highest_description})")
    print(f"Lowest expense: {currency} {lowest_expense:.2f} (Description: {lowest_description})")

    # Categorize expenses by size
    # Large (>20% of daily budget), Medium (5-20%), Small (<5%)
    large_count = 0
    medium_count = 0
    small_count = 0

    for expense in expenses:
        # Using daily budget as a reference for categories
        expense_percentage_of_daily_budget = (expense['amount'] / daily_budget) * 100
        if expense_percentage_of_daily_budget > 20:
            large_count += 1
        elif 5 <= expense_percentage_of_daily_budget <= 20:
            medium_count += 1
        else:
            small_count += 1

    print(f"\n📊 Expense Categories (relative to daily budget):")
    print(f"Large expenses (>20% daily budget): {large_count}")
    print(f"Medium expenses (5-20% daily budget): {medium_count}")
    print(f"Small expenses (<5% daily budget): {small_count}")

    # Smart Feedback
    print("\n💡 Smart Feedback:")
    if budget_used_percentage > 80:
        print("You are close to exceeding your monthly budget. Consider cutting back on non-essential spending.")
    elif budget_used_percentage > 50:
        print("You are managing your budget well, but keep an eye on your spending to stay on track.")
    else:
        print("Excellent job staying well within your budget! Keep up the great work.")

    if large_count > 0:
        print("You have a few large expenses. Review them to see if any can be reduced or optimized.")
    if remaining_budget < 0:
        print("Warning: You have exceeded your monthly budget. Time for a financial review!")

else:
    print("📝 No expenses entered yet. Add some to see the analysis!")
```

## Part 3: Multiple Savings Goals Tracker (15 minutes)

### 🎯 Goal Management with Loops

Extend your tracker to handle multiple savings goals:

```python
print("\n=== Savings Goals Tracker ===")

# Goal collection
goal_names = []
goal_targets = []
goal_current = []

# Ask how many goals they want to track
num_goals = 0
while True:
    try:
        num_goals_str = input("How many savings goals? (1-5): ")
        num_goals = int(num_goals_str)
        if 1 <= num_goals <= 5:
            break
        else:
            print("Please enter a number between 1 and 5.")
    except ValueError:
        print("Invalid input. Please enter a valid number!")

print(f"\nEntering {num_goals} savings goals:")

for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    goal_name = input(f"Enter name for goal {i+1}: ")
    goal_names.append(goal_name)

    target_amount = 0.0
    while True:
        try:
            target_amount_str = input(f"Enter target amount for '{goal_name}': ")
            target_amount = float(target_amount_str)
            if target_amount <= 0:
                print("Target amount must be a positive number. Please try again.")
            else:
                goal_targets.append(target_amount)
                break
        except ValueError:
            print("Invalid input. Please enter a numerical value for the target amount.")

    current_amount = 0.0
    while True:
        try:
            current_amount_str = input(f"Enter current amount saved for '{goal_name}': ")
            current_amount = float(current_amount_str)
            if current_amount < 0:
                print("Current amount cannot be negative. Please try again.")
            else:
                goal_current.append(current_amount)
                break
        except ValueError:
            print("Invalid input. Please enter a numerical value for the current amount.")

print(f"\n✅ Saved {len(goal_names)} goals!")
```

## Part 4: Simple Menu System (10 minutes)

### 🎮 Interactive Program Flow

Create a basic menu that lets users choose what to do:

```python
print("\n=== Interactive Finance Menu ===")

# Simple menu system
running = True

while running:
    print("\n" + "-"*30)
    print(f"💰 {name}'s Finance Tracker")
    # Display some current stats if available
    if len(expenses) > 0:
        # Ensure total_spent is available from previous executions or is initialized
        # Assuming total_spent is already a global variable updated by expense entry
        print(f"📊 Current: {len(expenses)} expenses, {currency} {total_spent:.2f} spent")
    if len(goal_names) > 0:
        # Assuming total_current_all_goals is available from goal analysis or is initialized
        # If not, it would need to be calculated here or passed around
        print(f"🎯 {len(goal_names)} savings goals actively tracked.")

    print("-"*30)
    print("1. View Expense Summary")
    print("2. View Goals Progress")
    print("3. Add New Expense")
    print("4. Budget Check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    if choice == "1":
        # Logic for viewing expense summary (from Part 2 analysis)
        print("\n--- Expense Summary ---")
        if len(expenses) > 0:
            print(f"Total expenses recorded: {len(expenses)}")
            print(f"Total amount spent: {currency} {total_spent:.2f}")
            average_expense = total_spent / len(expenses)
            print(f"Average expense: {currency} {average_expense:.2f}")
            print("\nDetailed Expenses:")
            for idx, exp in enumerate(expenses):
                print(f"{idx+1}. {exp['description']} ({exp['category']}): {currency} {exp['amount']:.2f}")
        else:
            print("No expenses to display yet.")

    elif choice == "2":
        # Logic for viewing goals progress (from Part 3 analysis)
        print("\n--- Goals Progress ---")
        if len(goal_names) > 0:
            print("\n🎯 Individual Goal Progress:")
            for i in range(len(goal_names)):
                goal_name = goal_names[i]
                target = goal_targets[i]
                current = goal_current[i]

                progress_percentage = (current / target) * 100 if target > 0 else 0
                amount_remaining = target - current

                print(f"\nGoal: {goal_name}")
                print(f"  Target: {currency} {target:.2f}")
                print(f"  Current: {currency} {current:.2f}")
                print(f"  Progress: {progress_percentage:.1f}%")
                if amount_remaining > 0:
                    print(f"  Amount Remaining: {currency} {amount_remaining:.2f}")
                else:
                    print("  Goal Achieved! 🎉")
        else:
            print("No savings goals to display yet.")

    elif choice == "3":
        # Logic for adding a single new expense (similar to Part 1 expense entry)
        print("\n--- Add New Expense ---")
        while True:
            expense_amount_str = input("Enter expense amount (or 'done' to cancel): ").lower()
            if expense_amount_str == 'done':
                print("Expense addition cancelled.")
                break

            try:
                amount = float(expense_amount_str)
                if amount <= 0:
                    print("Expense amount must be a positive number. Please try again.")
                    continue
            except ValueError:
                print("Invalid input. Please enter a numerical value for the amount.")
                continue

            description = input("Enter expense description: ")
            category = input("Enter expense category (e.g., Food, Transport, Utilities): ")

            expenses.append({
                'amount': amount,
                'description': description,
                'category': category
            })
            global total_spent # Declare as global to modify the variable from the outer scope
            total_spent += amount
            print(f"Added: {description} - {currency} {amount:.2f}")
            break # Exit the loop after successfully adding one expense

    elif choice == "4":
        # Logic for budget status check (from Part 2 analysis)
        print("\n--- Budget Check ---")
        if monthly_budget > 0:
            budget_used_percentage = (total_spent / monthly_budget) * 100
            remaining_budget = monthly_budget - total_spent
            print(f"Monthly Budget: {currency} {monthly_budget:.2f}")
            print(f"Total spent so far: {currency} {total_spent:.2f}")
            print(f"Monthly Budget Used: {budget_used_percentage:.1f}%")
            print(f"Remaining Monthly Budget: {currency} {remaining_budget:.2f}")

            if budget_used_percentage > 80:
                print("💡 You are close to exceeding your monthly budget. Consider cutting back on non-essential spending.")
            elif budget_used_percentage > 50:
                print("💡 You are managing your budget well, but keep an eye on your spending to stay on track.")
            else:
                print("💡 Excellent job staying well within your budget! Keep up the great work.")

            if remaining_budget < 0:
                print("⚠️ Warning: You have exceeded your monthly budget. Time for a financial review!")
        else:
            print("Monthly budget not set or is zero. Cannot perform budget check.")

    elif choice == "5":
        # Exit with summary
        running = False
        print(f"\nGoodbye {name}!")
        print(f"Final summary: {len(expenses)} expenses, {currency} {total_spent:.2f} total")
        print("🚀 Your finance tracker is growing stronger each week!")

    else:
        print("Invalid choice! Please choose 1-5.")

print("\n=== Finance Tracker Session Complete ===")
```
