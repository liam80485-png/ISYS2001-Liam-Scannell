## Part 2: Financial Decision Logic (15 minutes)

### Budget Limit Checker
Apply your if/else skills to real budget management:

````python
print("=== Budget Limit Checker ===")

# Get user's budget and expense
monthly_budget = float(input("Enter your monthly budget: $"))
expense = float(input("Enter expense amount: $"))


if expense > 0.50 * monthly_budget:
    print("Major purchase - think carefully")
elif expense > 0.25 * monthly_budget:
    print("Significant expense")
elif expense > 0.10 * monthly_budget:
    print("Moderate expense")
else:
    print("Minor expense - within budget")
````

### Savings Goal Progress
Use multiple conditions to track savings progress:

````python
print("=== Savings Goal Tracker ===")

goal_amount = float(input("Savings goal: $"))
current_savings = float(input("Current savings: $"))

# Calculate progress percentage
if goal_amount > 0:
    progress_percentage = (current_savings / goal_amount) * 100
else:
    progress_percentage = 0 # Handle case where goal_amount is 0 to avoid division by zero

if progress_percentage >= 100:
    print("Goal achieved! Consider a new goal.")
elif progress_percentage >= 75:
    print("Almost there! You're doing great.")
elif progress_percentage >= 50:
    print("Halfway there - keep it up!")
elif progress_percentage >= 25:
    print("Good progress, stay consistent")
else:
    print("Just getting started - you've got this!")
````

## Part 3: Smart Purchase Advisor (20 minutes)

### The Challenge
Build a program that helps users make smart purchase decisions by considering multiple factors.

### Requirements

Your purchase advisor should:

1. Get purchase amount and user's available money
2. Check if they can afford it (basic affordability)
3. Give different advice based on purchase size
4. Consider their financial situation

### Starter Framework

````python
print("Welcome to the Smart Purchase Advisor!")
print("This program will help you decide if a purchase is a good idea.\n")

# Get user information
purchase_amount = float(input("Enter the price of the item you want to buy: "))
available_money = float(input("Enter how much money you currently have available: "))

# Check affordability
if purchase_amount > available_money:
    print("\n You cannot afford this purchase right now.")
    print("Advice: Consider saving more money before buying this item.")

else:
    remaining_money = available_money - purchase_amount
    print("\n You can afford this purchase.")

    # Advice based on purchase size
    if purchase_amount < 50:
        print("This is a small purchase. It should be fine if you really need it.")

    elif purchase_amount < 200:
        print("This is a medium purchase.")
        print("Advice: Make sure it's something useful before spending the money.")

    else:
        print("This is a large purchase.")
        print("Advice: Think carefully and consider waiting 24 hours before buying.")

    # Consider financial situation
    print(f"\nAfter buying this item, you would have ${remaining_money:.2f} left.")

    if remaining_money < 20:
        print(" Warning: This would leave you with very little money.")
        print("You might want to reconsider the purchase.")

    else:
        print(" Your finances would still be in a comfortable position.")
````

### Sample Decision Logic to Implement

- **Can't afford**: "Save up first - you need $X more"
- **Can afford but expensive**: "You can afford it, but it's Y% of your available money. Consider waiting."
- **Reasonable purchase**: "This seems like a reasonable purchase for your budget"
- **Small purchase**: "This is a minor expense - go for it!"

## Part 4: Building Your Finance Tracker Foundation (15 minutes)

### Start Your Semester Project

Create a new notebook called "Personal_Finance_Tracker" and build the foundation:

````python
print("=== Personal Finance Tracker ===")


name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")


print("\n--- Expense Entry ---")

expense_amount = float(input("Enter the expense amount: $"))
expense_description = input("Enter a description for the expense: ")


expense_category = ""
if expense_amount >= 100:
    expense_category = "Major Expense"
    print(f"Feedback: '{expense_description}' is a {expense_category}. This is a significant amount, consider if it's essential.")
elif expense_amount >= 50:
    expense_category = "Moderate Expense"
    print(f"Feedback: '{expense_description}' is a {expense_category}. Keep an eye on these to stay within budget.")
elif expense_amount >= 10:
    expense_category = "Minor Expense"
    print(f"Feedback: '{expense_description}' is a {expense_category}. Reasonable, but small expenses can add up!")
else:
    expense_category = "Small Purchase"
    print(f"Feedback: '{expense_description}' is a {expense_category}. Likely not a budget concern, but track it anyway.")

print(f"Expense recorded: ${expense_amount:.2f} for '{expense_description}' (Category: {expense_category})")
````
