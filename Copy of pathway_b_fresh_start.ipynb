### Step 2: Initialize Your Personal Finance Tracker

````python
# ============================================================================
# INITIALIZATION: CREATE PERSONALIZED FINANCE TRACKER
# ============================================================================
from datetime import datetime
import pyinputplus as pyip

def create_tracker():
    print("💰 FINANCE TRACKER INITIALISATION")
    print("=" * 45)

    # --- USER INPUT ---
    name = input("Enter your name: ")
    email = input("Enter your email: ")
    currency = input("Currency (default AUD): ") or "AUD"

    monthly_budget = pyip.inputFloat("Monthly budget: $", min=0)
    savings_goal = pyip.inputFloat("Monthly savings goal: $", min=0)
    emergency_goal = pyip.inputFloat("Emergency fund target: $", min=0)

    # --- CREATE TRACKER ---
    tracker = {
        "metadata": {
            "app_version": "3.0",
            "created_at": datetime.now().isoformat(),
            "last_updated": datetime.now().isoformat()
        },

        "user": {
            "name": name,
            "email": email,
            "currency": currency
        },

        "transactions": {
            "expenses": {},
            "income": {}
        },

        "categories": {
            "Housing": {"type": "expense", "budget": 0.30},
            "Food": {"type": "expense", "budget": 0.25},
            "Transport": {"type": "expense", "budget": 0.15},
            "Entertainment": {"type": "expense", "budget": 0.10},
            "Savings": {"type": "expense", "budget": 0.20},
            "Salary": {"type": "income"}
        },

        "accounts": {
            "ACC001": {
                "name": "Main Account",
                "balance": 0.0
            }
        },

        "budgets": {
            "monthly_limit": monthly_budget
        },

        "goals": {
            "monthly_savings": {
                "target": savings_goal,
                "current": 0
            },
            "emergency_fund": {
                "target": emergency_goal,
                "current": 0
            }
        },

        "analytics": {
            "last_transaction_id": 0,
            "total_income": 0.0,
            "total_expenses": 0.0
        }
    }

    # --- CONFIRMATION OUTPUT ---
    print("\n✅ Tracker Initialised Successfully!")
    print(f"👤 User: {name}")
    print(f"💵 Budget: {currency} {monthly_budget:.2f}")
    print(f"🎯 Savings Goal: {currency} {savings_goal:.2f}")
    print(f"🛡️ Emergency Fund Target: {currency} {emergency_goal:.2f}")

    return tracker


# --- INITIALISE ---
finance_tracker = create_tracker()

print("\n🎉 Your finance tracker is ready!")
````

## 🔧 Phase 2: Core Functionality Implementation (45 minutes)

### Step 3: Essential Expense Management Functions

Build the core functions using dictionary best practices:

````python
# ============================================================================
# CORE FUNCTIONS: EXPENSE MANAGEMENT WITH DICTIONARY POWER
# ============================================================================
from datetime import datetime
import pyinputplus as pyip

# 🔢 Generate unique expense ID
def generate_expense_id(tracker):
    tracker["analytics"]["last_transaction_id"] += 1
    return f"EXP{tracker['analytics']['last_transaction_id']:03d}"


# ➕ Add Expense (main function)
def add_expense(tracker):
    print("\n➕ ADD NEW EXPENSE")
    print("-" * 30)

    # --- INPUT ---
    amount = pyip.inputFloat("Amount: $", min=0)
    description = input("Description: ")

    # Show available categories
    print("\nAvailable categories:")
    for cat, data in tracker["categories"].items():
        if data["type"] == "expense":
            print(f"- {cat}")

    category = input("Category: ")

    # --- VALIDATION ---
    if category not in tracker["categories"] or tracker["categories"][category]["type"] != "expense":
        print("❌ Invalid category.")
        return

    # --- CREATE EXPENSE ---
    exp_id = generate_expense_id(tracker)

    tracker["transactions"]["expenses"][exp_id] = {
        "amount": amount,
        "description": description,
        "category": category,
        "date": datetime.now().strftime("%Y-%m-%d")
    }

    # --- UPDATE ANALYTICS ---
    tracker["analytics"]["total_expenses"] += amount

    print(f"✅ Expense {exp_id} added successfully!")


# 📄 View All Expenses
def view_expenses(tracker):
    print("\n📄 ALL EXPENSES")
    print("-" * 40)

    expenses = tracker["transactions"]["expenses"]

    if not expenses:
        print("No expenses recorded.")
        return

    for exp_id, exp in expenses.items():
        print(f"{exp_id} | ${exp['amount']:.2f} | {exp['category']} | {exp['description']} | {exp['date']}")


# ❌ Delete Expense
def delete_expense(tracker):
    print("\n❌ DELETE EXPENSE")
    print("-" * 30)

    exp_id = input("Enter Expense ID (e.g. EXP001): ")

    if exp_id not in tracker["transactions"]["expenses"]:
        print("❌ Expense not found.")
        return

    # Adjust analytics before deleting
    amount = tracker["transactions"]["expenses"][exp_id]["amount"]
    tracker["analytics"]["total_expenses"] -= amount

    del tracker["transactions"]["expenses"][exp_id]

    print(f"✅ Expense {exp_id} deleted.")


# 🔍 Search Expenses by Category
def search_by_category(tracker):
    print("\n🔍 SEARCH BY CATEGORY")
    print("-" * 30)

    category = input("Enter category: ")

    found = False
    for exp_id, exp in tracker["transactions"]["expenses"].items():
        if exp["category"] == category:
            print(f"{exp_id} | ${exp['amount']:.2f} | {exp['description']} | {exp['date']}")
            found = True

    if not found:
        print("No matching expenses found.")


# 📊 Expense Summary
def expense_summary(tracker):
    print("\n📊 EXPENSE SUMMARY")
    print("-" * 30)

    total = tracker["analytics"]["total_expenses"]
    count = len(tracker["transactions"]["expenses"])

    print(f"Total Expenses: ${total:.2f}")
    print(f"Number of Transactions: {count}")

    # Category breakdown (calculated dynamically = safer)
    category_totals = {}

    for exp in tracker["transactions"]["expenses"].values():
        cat = exp["category"]
        category_totals[cat] = category_totals.get(cat, 0) + exp["amount"]

    print("\nBy Category:")
    for cat, amount in category_totals.items():
        print(f"- {cat}: ${amount:.2f}")
````
