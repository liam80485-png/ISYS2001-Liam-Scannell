## Step 5 – Write the Code

Translate your pseudocode into Python code. Below is one example solution:

````python
# 1. Start the program
print("Welcome to the Finance Tracker App!")

# 2. Prompt for user's name
name = input("Enter your name: ")

# 3. Prompt for preferred theme color
theme_color = input("Enter your preferred theme color: ")

# 4. Prompt for primary currency
currency = input("Enter your primary currency (e.g., USD, AUD, EUR): ")

# 5. Prompt for main savings goal
savings_goal = input("Enter your main savings goal: ")

# 6. Prompt for notification preference
notification_preference = input("Do you want to receive daily notifications? (yes/no): ")

# 7. Prompt for daily budget
daily_budget = input("What is your preferred daily budget?: ")

# 8. Construct welcome and summary message
summary_message = (
    f"\nHello {name}!\n"
    f"Your app theme color is set to {theme_color}.\n"
    f"Your primary currency is {currency}.\n"
    f"Your main savings goal is: {savings_goal}.\n"
    f"Daily notifications are turned {notification_preference}.\n"
    f"Your preferred daily budget is: {daily_budget} {currency}.\n"
    f"You're all set to start tracking your finances!"
)

# 9. Print the summary message
print(summary_message)

# 10. End the program
print("\nSetup complete. Thank you for using the Finance Tracker App!")
````
